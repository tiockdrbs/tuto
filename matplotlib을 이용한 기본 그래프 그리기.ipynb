### 그래프 선 색깔 바꾸기
r = 빨간색, g = 초록색, b = 파란색, k = 검정색,
c = 청록색, m = 자주색, y = 노란색, w = 흰색

```python
plt.figure(figsize=(20,5))

plt.plot(town,baby_0,'r',label='0세')
plt.plot(town,child,'g',label='8세')

plt.title('0세, 8세 동네별 인구수',fontsize=20)
plt.xlabel('서울특별시 종로구 동',fontsize=15)
plt.ylabel('연령 별 인구 수')

plt.legend()
plt.show()
```

### 그래프 선 모양 바꾸기
'-' = 실선, '--' = 대시선, '-.' = 대시-점 선, ':' = 점선

```python
plt.figure(figsize=(20,5))
plt.plot(town,baby_0,'r--',label='0세')
plt.plot(town,child,'g-.',label='8세')

plt.title('0세, 8세 동네별 인구수',fontsize=20)
plt.xlabel('서울특별시 종로구 동',fontsize=15)
plt.ylabel('연령 별 인구 수')


plt.legend()
plt.show()
```

### 그래프 마커 추가하기

'p' = 오각형, 'h1'= 육각형, '+' = 더하기, 'x' = X모양

```python
plt.figure(figsize=(20,5))
plt.plot(town,baby_0,'ro--',label='0세')
plt.plot(town,child,'gx-.',label='8세')

plt.title('0세, 8세 동네별 인구수',fontsize=20)
plt.xlabel('서울특별시 종로구 동',fontsize=15)
plt.ylabel('연령 별 인구 수')

plt.legend()
plt.show()
```

### 0~6세 누적분포

```python
#2020년07월_계_0세
#plt.bar(index, child, color='c', bottom=baby, label='8세')
plt.figure(figsize=(20,5))

for i in range(0,7):
    if i==0:
        plt.bar(town, age['2020년07월_계_'+str(i)+'세'],label=str(i)+'세')
    else:
        plt.bar(town, age['2020년07월_계_'+str(i)+'세'],bottom=age['2020년07월_계_'+str(i-1)+'세'],label=str(i)+'세')
    
plt.title('0세~6세 동네별 인구수',fontsize=20)
plt.xlabel('서울특별시 종로구 동네')
plt.ylabel('연령 별 인구 수')
plt.legend()
    
plt.show()
```

```python
#2020년07월_계_0세
#plt.bar(index, child, color='c', bottom=baby, label='8세')
index= np.arange(len(town))
bar_width = 0.35 

plt.figure(figsize=(20,5))

for i in range(0,10):
    if i==0:
        plt.bar(index, age['2020년07월_계_'+str(i)+'세'],bar_width,label=str(i)+'세')
    elif i>=1 and i<=4:
        plt.bar(index, age['2020년07월_계_'+str(i)+'세'],bar_width,bottom=age['2020년07월_계_'+str(i-1)+'세'],label=str(i)+'세')
    elif i==5:
        plt.bar(index+bar_width, age['2020년07월_계_'+str(i)+'세'],bar_width,label=str(i)+'세')
    else:
        plt.bar(index+bar_width, age['2020년07월_계_'+str(i)+'세'],bar_width,bottom=age['2020년07월_계_'+str(i-1)+'세'],label=str(i)+'세')
    

plt.title('0세~4세 및 5~9세 동네별 인구 수',fontsize=20)
plt.xlabel('서울특별시 종로구 동네')
plt.ylabel('인구 수')
plt.xticks(index, town, fontsize=12)
plt.legend()
    
plt.show()
```

## 원 그래프 그리기

```python
age_1 = age.set_index('행정구역')
pyeongchang=age_1.loc['평창동']
```

```python
name_chart = ['0세','1세','2세','3세','4세']

values=[]
for i in range(0,5):
    values.append(pyeongchang['2020년07월_계_'+str(i)+'세'])
    
color_chart=['r','b','y','c','g']
explode_chart=(0,0,0,0.1,0.1) #0일 때, 그래프가 붙어있고, 0.1만큼 떨어지게 하는 것

plt.title('평창동의 0~4세 비율',fontsize=20)
plt.pie(values, explode=explode_chart,labels=name_chart,colors=color_chart, autopct='%1.2f%%')
plt.show()
```
